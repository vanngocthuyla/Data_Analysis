```
# Load the data
data = read.table("15_Markers.txt", header = TRUE)
gene_names = data[2:nrow(data), 1]
groups = data[1, 2:ncol(data)]
expressions = as.matrix(data[2:nrow(data), 2:ncol(data)])
class(expressions) = "numeric"  # Convert strings to numeric values
row.names(expressions) = gene_names
expressions = t(expressions)

head(expressions)
```

```
pca_obj = prcomp(expressions)
summary(pca_obj)
groups = t(groups)
colnames(groups) = "Group"
```

### Plotting

```
library(ggfortify)
library(ggplot2)
autoplot(pca_obj, data=groups, colour="Group", 
	 label=TRUE, label.size=2, shape=FALSE,
	 loadings = TRUE, loadings.colour = 'blue',
         loadings.label = TRUE, loadings.label.size = 3, 
				 frame = "TRUE")
```

Change the type of frame: 

```
autoplot(pca_obj, data=groups, colour="Group", 
	 label=TRUE, label.size=2, shape=FALSE,
	 loadings = TRUE, loadings.colour = 'blue',
         loadings.label = TRUE, loadings.label.size = 3, 
				 frame = "TRUE", frame.type = 'norm')
```

Or plotting without eigenvectors:

```
autoplot(pca_obj, data=groups, colour="Group", 
	 label=TRUE, label.size=2, shape=FALSE,frame = "TRUE", frame.type = 'norm')
```
